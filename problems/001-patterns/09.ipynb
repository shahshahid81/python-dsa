Steps to print any pattern
* For the outer loop, count the number of rows (lines).
* For the inner loop, focus on the columns(character to print) and somehow relate the column to the row.
* Print the character inside the inner loop.
* Observe symmetry. (This step is optional).

![Pattern 9](../../images/001-patterns/9.png)

```python
def print_pattern(n):
    # n = 5
    # 4 space 1 star 4 space
    # 3 space 3 star 3 space
    # 2 space 5 star 2 space
    # 1 space 7 star 1 space
    # 0 space 9 star 0 space
    for i in range(n):
        space = " " * (n - 1 - i)
        row = space + "*" * (i * 2 + 1) + space
        print(row)

    # n = 5
    # 0 space 9 star 0 space
    # 1 space 7 star 1 space
    # 2 space 5 star 2 space
    # 3 space 3 star 3 space
    # 4 space 1 star 4 space
    for i in range(n):
        space = " " * i
        row = space + "*" * ((n - i) * 2 - 1) + space
        print(row)


print_pattern(5)
```
