Steps to print any pattern
* For the outer loop, count the number of rows (lines).
* For the inner loop, focus on the columns(character to print) and somehow relate the column to the row.
* Print the character inside the inner loop.
* Observe symmetry. (This step is optional).

![Pattern 17](../../images/001-patterns/17.png)

```python
def print_pattern(n):
    # n = 4
    # 3 space 1 char 0 reverse char 3 space
    # 2 space 2 char 1 reverse char 2 space
    # 1 space 3 char 2 reverse char 1 space
    # 0 space 4 char 3 reverse char 0 space

    start = ord("A")
    for i in range(1, n + 1):
        spaces = " " * (n - i)
        first_half = "".join(chr(start + j) for j in range(i))
        second_half = "".join(chr(start + j - 1) for j in range(i - 1, 0, -1))
        print(spaces + first_half + second_half + spaces)


print_pattern(4)
```
