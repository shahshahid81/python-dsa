Steps to print any pattern
* For the outer loop, count the number of rows (lines).
* For the inner loop, focus on the columns(character to print) and somehow relate the column to the row.
* Print the character inside the inner loop.
* Observe symmetry. (This step is optional).

```python
def print_pattern_row(i, n):
    stars = "*" * i
    spaces = " " * (n - i) * 2
    print(stars + spaces + stars)


def print_pattern(n):
    # n = 4
    # 4 stars
    # 1 star 2 space 1 star
    # 1 star 2 space 1 star
    # 4 stars

    # n = 5
    # 5 stars
    # 1 star 3 space 1 star
    # 1 star 3 space 1 star
    # 1 star 3 space 1 star
    # 5 stars
    full_row = "*" * n
    print(full_row)
    for _ in range(1, n - 1):
        spaces = " " * (n - 2)
        print("*" + spaces + "*")
    print(full_row)


print_pattern(4)
```
