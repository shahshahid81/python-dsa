Steps to print any pattern
* For the outer loop, count the number of rows (lines).
* For the inner loop, focus on the columns(character to print) and somehow relate the column to the row.
* Print the character inside the inner loop.
* Observe symmetry. (This step is optional).

![Pattern 12](../../images/001-patterns/12.png)

```python
def print_pattern(n):
    # n = 4
    # 1 num 6 space 1 num reverse
    # 2 num 4 space 2 num reverse
    # 3 num 2 space 3 num reverse
    # 4 num 0 space 4 num reverse
    for i in range(1, n + 1):
        left = "".join(str(j) for j in range(1, i + 1))
        spaces = " " * ((n - i) * 2)
        right = "".join(str(j) for j in range(i, 0, -1))
        print(left + spaces + right)


print_pattern(4)
```
