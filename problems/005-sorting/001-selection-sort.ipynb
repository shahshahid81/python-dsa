```python
def selection_sort(lst: list[int]) -> list[int]:
    n = len(lst)
    list_copy = lst.copy()

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if list_copy[j] < list_copy[min_index]:
                min_index = j

        list_copy[i], list_copy[min_index] = list_copy[min_index], list_copy[i]

    return list_copy


print(selection_sort([1, 5, 3, 4]))
print(selection_sort([13, 46, 24, 52, 20, 9]))
print(selection_sort([5, 4, 3, 2, 1]))
print(selection_sort([1, 2, 3, 4, 5]))
print(selection_sort([]))
```
