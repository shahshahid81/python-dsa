```python
def partition(lst, low, high):
    # i marks the next position where an element smaller than the pivot should be placed.
    i = low
    # Store the pivot value for comparison.
    pivot = lst[high]

    # j scans every element before the pivot.
    for j in range(low, high):
        # Move elements smaller than the pivot to the left partition.
        # i advances to the next available position in that partition.
        if lst[j] < pivot:
            lst[i], lst[j] = lst[j], lst[i]
            i += 1

    # Place the pivot between the two partitions.
    # All elements to the left are < pivot, and all elements to the right are >= pivot.
    lst[high], lst[i] = lst[i], lst[high]

    return i


def quick_sort_helper(lst, low, high):
    if low >= high:
        return lst

    pivot_index = partition(lst, low, high)

    quick_sort_helper(lst, low, pivot_index - 1)
    quick_sort_helper(lst, pivot_index + 1, high)

    return lst


def quick_sort(lst):
    return quick_sort_helper(lst.copy(), 0, len(lst) - 1)


print(quick_sort([1, 5, 3, 4]))
print(quick_sort([13, 46, 24, 52, 20, 9]))
print(quick_sort([5, 4, 3, 2, 1]))
print(quick_sort([1, 2, 3, 4, 5]))
print(quick_sort([]))
```
