```python
def merge(left: list[int], right: list[int]) -> list[int]:
    result = []
    i = 0
    j = 0
    m = len(left)
    n = len(right)
    while i < m and j < n:
        if left[i] > right[j]:
            result.append(right[j])
            j += 1
        else:
            result.append(left[i])
            i += 1

    while i < m:
        result.append(left[i])
        i += 1

    while j < n:
        result.append(right[j])
        j += 1

    return result


def merge_sort_helper(lst: list[int], low: int, high: int):
    if low == high:
        return [lst[low]]

    mid = low + (high - low) // 2
    left = merge_sort_helper(lst, low, mid)
    right = merge_sort_helper(lst, mid + 1, high)

    return merge(left, right)


def merge_sort(lst: list[int]) -> list[int]:
    if not lst:
        return []
    n = len(lst)
    result = lst.copy()
    result = merge_sort_helper(result, 0, n - 1)
    return result


print(merge_sort([1, 5, 3, 4]))
print(merge_sort([13, 46, 24, 52, 20, 9]))
print(merge_sort([5, 4, 3, 2, 1]))
print(merge_sort([1, 2, 3, 4, 5]))
print(merge_sort([]))
```
