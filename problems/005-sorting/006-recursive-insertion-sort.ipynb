```python
def insertion_sort_helper(lst, i):
    n = len(lst)
    if i >= n:
        return lst

    j = i
    while j > 0 and lst[j] < lst[j - 1]:
        lst[j], lst[j - 1] = lst[j - 1], lst[j]
        j -= 1

    return insertion_sort_helper(lst, i + 1)


def insertion_sort(lst):
    return insertion_sort_helper(lst.copy(), 0)


print(insertion_sort([1, 5, 3, 4]))
print(insertion_sort([13, 46, 24, 52, 20, 9]))
print(insertion_sort([5, 4, 3, 2, 1]))
print(insertion_sort([1, 2, 3, 4, 5]))
print(insertion_sort([]))
```
