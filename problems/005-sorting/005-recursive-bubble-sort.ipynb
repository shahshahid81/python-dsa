```python
def bubble_sort_helper(lst, n):
    if n == 0:
        return lst

    did_swap = False
    for i in range(n - 1):
        if lst[i] > lst[i + 1]:
            lst[i], lst[i + 1] = lst[i + 1], lst[i]
            did_swap = True

    return bubble_sort_helper(lst, n - 1) if did_swap else lst


def bubble_sort(lst):
    n = len(lst)
    return bubble_sort_helper(lst.copy(), n)


print(bubble_sort([1, 5, 3, 4]))
print(bubble_sort([13, 46, 24, 52, 20, 9]))
print(bubble_sort([5, 4, 3, 2, 1]))
print(bubble_sort([1, 2, 3, 4, 5]))
print(bubble_sort([]))
```
