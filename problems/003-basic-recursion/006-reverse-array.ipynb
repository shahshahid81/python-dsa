```python
def reverse_array_helper(arr: list[int], low: int, high: int) -> list[int]:
    if low >= high:
        return arr
    arr[high], arr[low] = arr[low], arr[high]
    reverse_array_helper(arr, low + 1, high - 1)


def reverse_array(arr: list[int]) -> list[int]:
    reverse_array_helper(arr, 0, len(arr) - 1)
    return arr


print(reverse_array([1, 2, 3]))
print(reverse_array([1, 2, 3, 4]))
print(reverse_array([]))
```
