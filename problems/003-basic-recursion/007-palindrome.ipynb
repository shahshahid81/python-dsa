```python
def is_palindrome_helper(value: str, low: int, high: int) -> bool:
    if low >= high:
        return True
    if value[low] == value[high]:
        return is_palindrome_helper(value, low + 1, high - 1)
    return False


def is_palindrome(value: str) -> bool:
    return is_palindrome_helper(value, 0, len(value) - 1)


print(is_palindrome("MADAM"))
print(is_palindrome("ABBY"))
print(is_palindrome("A"))
```
