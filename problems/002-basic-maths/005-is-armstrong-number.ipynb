```python
import math


def number_of_digits(n):
    if n == 0:
        return 1
    return math.floor(math.log10(abs(n))) + 1


def is_armstrong_number(n):
    if n < 0:
        return False
    x = n
    digits = number_of_digits(n)
    sum_n = 0
    while n != 0:
        sum_n = sum_n + ((n % 10) ** digits)
        n //= 10
    return x == sum_n


print(is_armstrong_number(153))
print(is_armstrong_number(371))
```
