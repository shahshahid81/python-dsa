```python
import math


def count_all_digits(n: int) -> int:
    n = abs(n)
    if n == 0:
        return 1
    # Each power of 10 adds one more digit.
    # log10(n) tells us which powers of 10 the number falls between.
    # Taking the floor gives the lower power, and adding 1 gives the digit count.
    #
    # Examples:
    # 7    -> between 10^0 and 10^1 -> floor(log10(7)) = 0 -> 1 digit
    # 42   -> between 10^1 and 10^2 -> floor(log10(42)) = 1 -> 2 digits
    # 999  -> between 10^2 and 10^3 -> floor(log10(999)) = 2 -> 3 digits
    # 1000 -> 10^3                  -> floor(log10(1000)) = 3 -> 4 digits
    return math.floor(math.log10(n)) + 1


# def count_all_digits(n: int) -> int:
#     n = abs(n)
#     if n == 0:
#         return 1
#     count = 0
#     while n != 0:
#         n //= 10
#         count += 1
#     return count


print(count_all_digits(12345))
print(count_all_digits(7789))
print(count_all_digits(-1))
print(count_all_digits(-7))
print(count_all_digits(0))
print(count_all_digits(-2147483648))
```
