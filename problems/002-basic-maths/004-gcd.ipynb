```python
# def gcd(a, b):
#    if any one parameter is 0, return the positive version of the other parameter
#     if a == 0:
#         return abs(b)

#     if b == 0:
#         return abs(a)

#     result = 1
#     min_number = min(a, b)
#     for i in range(2, min_number + 1):
#         if a % i == 0 and b % i == 0:
#             result = i
#     return result


def gcd(a, b):
    # Euclidean Algorithm
    #
    # Algorithm:
    # 1. Divide the larger number by the smaller.
    # 2. Replace the larger number with the smaller number.
    # 3. Replace the smaller number with the remainder.
    # 4. Repeat until the remainder becomes 0.
    # 5. The last non-zero number is the GCD.
    #
    # Why does this work?
    #
    # The GCD does not change if we replace the larger number
    # with the remainder after dividing it by the smaller.
    #
    #     gcd(a, b) = gcd(b, a % b)
    #
    # Example:
    #
    #     gcd(45, 10)
    #
    #     45 = 10 * 4 + 5
    #
    # If a number divides both 45 and 10, it must also divide:
    #
    #     45 - (10 * 4) = 5
    #
    # Likewise, if a number divides both 10 and 5, it must also divide:
    #
    #     (10 * 4) + 5 = 45
    #
    # Therefore, (45, 10) and (10, 5) have the same common divisors,
    # so they have the same GCD.
    #
    # Reduction:
    #
    #     gcd(45, 10)
    #     → gcd(10, 5)
    #     → gcd(5, 0)
    #     → 5
    #
    # The order of the inputs doesn't matter.
    # If a < b, then a % b == a, so the first iteration swaps them.
    #
    # Example:
    #
    #     gcd(10, 45)
    #     → gcd(45, 10)
    #     → gcd(10, 5)
    #     → gcd(5, 0)
    #     → 5
    #
    # Every pair of integers has a GCD.
    # If the numbers have no common factor greater than 1,
    # their GCD is 1 (they are coprime).
    #
    # Examples:
    #
    #     gcd(8, 15) = 1
    #     gcd(0, 5) = 5
    #     gcd(0, 0) = 0  # Convention used by Python's math.gcd()

    a, b = abs(a), abs(b)

    while b:
        a, b = b, a % b

    return a

print(gcd(9, 12))
print(gcd(20, 15))
```
