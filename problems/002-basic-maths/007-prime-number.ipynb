```python
import math

# def check_prime_number(n):
#     if n <= 1:
#         return False

#     for i in range(2, n):
#         if n % i == 0:
#             return False
#     return True

def check_prime_number(n):
    if n <= 1:
        return False

    # Divisors always occur in pairs.
    # If i divides n, then n // i also divides n.
    # Therefore, we only need to iterate up to and including √n.
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False

    return True


print(check_prime_number(2))
print(check_prime_number(10))
print(check_prime_number(13))
print(check_prime_number(15))
```
