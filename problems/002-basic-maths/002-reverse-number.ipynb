```python
import math


class Solution:
    def reverse(self, n: int) -> int:
        is_negative = True if n < 0 else False
        n = abs(n)
        rev = 0
        while n != 0:
            rev = rev * 10 + n % 10
            n //= 10
        result = rev * -1 if is_negative else rev
        if result > math.pow(2, 31) - 1 or result < math.pow(-2, 31):
            return 0
        return result


s = Solution()
print(s.reverse(12345))
print(s.reverse(7789))
print(s.reverse(-123))
print(s.reverse(0))
```
