```python
import math

# Brute Force Approach - O(n)
# def print_all_divisors(n: int) -> None:
#     divisors = [i for i in range(1, n + 1) if n % i == 0]
#     print(divisors)


# Optimized Approach - O(√n)
def print_all_divisors(n: int) -> None:
    divisors = []

    # Divisors always occur in pairs.
    # If i divides n, then n // i also divides n.
    # Therefore, we only need to iterate up to √n.
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % i == 0:
            divisors.append(i)

            # Avoid adding the square root twice for perfect squares
            # (e.g., 36 = 6 × 6).
            if i != n // i:
                divisors.append(n // i)

    print(sorted(divisors))


print_all_divisors(12)
print_all_divisors(36)
```
